```python
people = {
    'first': ['Corey', 'Jane', 'Hale', 'Adam'],
    'last': ['Schafner', 'Doe', 'Doe', 'Doe'],
    'email': ['Corey@gmail.com', 'Jane@gmail.com', 'Hale@gmail.com', 'Adam@gmail.com']
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(people)

df
```

## sort_values

```python
df.sort_values(by='last') # Ascending order
```

```python
df.sort_values(by='last', ascending=False) # Descending order
```

```python
df.sort_values(by=['last', 'first'], ascending=False) # Multiple fields sorting | Descending order
```

```python
df.sort_values(by=['last', 'first'], ascending=[False, True], inplace=True) # last => False | first => True

df
```

## sort_index()

```python
df.sort_index()
```

```python
df['last'].sort_values()
```
